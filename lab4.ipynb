## Simple Linear Regression

## Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

## Step 2. Import the dataset

```python
df = pd.read_csv('50_Startups.csv')
```

```python
df
```

## Step 3 . Check the State Column

```python
df["State"].value_counts()
```

## Step 4 . Splitting dataset in to input and output

```python
#splitting dataset in to input and output
X = df.iloc[:, :-1]
X2= X.copy()
y = df.iloc[:, -1]
```

## Step 5 . Convert state Column into Numeric Column

## Step 5.1 .  Perform Transformation

```python
# using dummies
sd = pd.get_dummies(df['State'], prefix='State', drop_first=True)
X = X.drop('State', axis=1)
X = pd.concat([X, sd], axis=1)
X
```

```python
encoder = OneHotEncoder(handle_unknown='ignore', drop='first')
cat_encoder = encoder.fit_transform(X2[['State']])
cat_encodede_df = pd.DataFrame(cat_encoder.toarray(), columns=encoder.get_feature_names_out(['State']))
cat_encodede_df

X2.drop('State', axis=1, inplace=True)
X2 = pd.concat([X2, cat_encodede_df], axis=1)
X2
```

## Step 6 . Dummy variable trap

```python
# Already Performed using | drop_first =True
```

## Step 7 Splitting dataset in to Train and Test

```python
x_train, x_test , y_train, y_test = train_test_split(X2, y, test_size=0.2, random_state=40)
```

## Step 8  Import LinearRegression model from linear_model family 

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

## Step 9  Fit the data

```python
model.fit(x_train, y_train)
```

## Step 10  Predict the data

```python
y_pred = model.predict(x_test)
y_pred
```

## Step 11  Display Result

```python
# y_test and y_predict with some beautification
result_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
# result_df.reset_index(drop=True, inplace=True)
result_df
```

## RSS

```python
#RSS

rss = np.sum((y_test - y_pred) ** 2)
float(rss)
```

## R Square

```python
# R square
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
r2
```

## Now use Polynomial Regression on Position_Salaries dataset

```python
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

df2 = pd.read_csv('Position_Salaries.csv')
X_poly = df2[["Level"]]
y_poly = df2["Salary"]

poly = PolynomialFeatures(degree=4, include_bias=False)
X_poly_features = poly.fit_transform(X_poly)

poly_reg = LinearRegression()
poly_reg.fit(X_poly_features, y_poly)
val = 6.7
sample_level = np.array([[val]])
sample_pred = poly_reg.predict(poly.transform(pd.DataFrame(sample_level, columns=["Level"])))

level_min = X_poly["Level"].min()
level_max = X_poly["Level"].max()
level_grid = np.arange(level_min, level_max + 0.1, 0.1).reshape(-1, 1)
level_grid_pred = poly_reg.predict(poly.transform(pd.DataFrame(level_grid, columns=["Level"])))

print(f"Predicted salary for level {val}: {sample_pred[0]:.2f}")

plt.scatter(X_poly, y_poly, color="blue", label="Actual")
plt.plot(level_grid, level_grid_pred, color="red", label="Polynomial fit (deg=4)")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.title("Polynomial Regression on Position Salaries")
plt.legend()
plt.show()
```
